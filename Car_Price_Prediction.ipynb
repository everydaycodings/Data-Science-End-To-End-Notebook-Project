```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
data = pd.read_csv("Datasets/car_price_prediction/quikr_car.csv")
data
```

```python
data.info()
```

```python
data = data[data["year"].str.isnumeric()]
data["year"] = data["year"].astype("int")
```

```python
data = data[data["Price"] != "Ask For Price"]
data["Price"] = data["Price"].str.replace(",", "").astype("int")
data
```

```python
data["kms_driven"] = data["kms_driven"].str.split(" ").apply(lambda x: x[0]).str.replace(",", "")
data = data[data["kms_driven"].str.isnumeric()]
data["kms_driven"] = data["kms_driven"].astype("int")
data.info()
```

```python
data = data[~data["fuel_type"].isna()]
data
```

```python
data["name"] = data["name"].str.split(" ").apply(lambda x: x[0:3]).str.join(" ")
data
```

```python
data = data[data["Price"] < 6e6]
```

```python
data = data.reset_index(drop=True)
data
```

```python
x = data.drop(columns="Price") # Features Column (all columns are features except Target)
y = data["Price"] # Lable is target column
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[["name", "company", "fuel_type"]])
```

```python
column_tranformer = make_column_transformer((OneHotEncoder(categories=ohe.categories_), ["name", "company", "fuel_type"]),
                                            remainder="passthrough")
```

scores = []

for i in range (1000):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=i)
    lr = LinearRegression()
    pipe = make_pipeline(column_tranformer, lr)
    pipe.fit(x_train, y_train)
    y_pre = pipe.predict(x_test)
    scores.append(r2_score(y_test, y_pre))

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=661)
lr = LinearRegression()
pipe = make_pipeline(column_tranformer, lr)
pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)
r2_score(y_test, y_pred)
```
