```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
data = pd.read_csv("Datasets/Football_Club_Market_Value_2021/club.csv")
data.head()
```

```python
data.info()
```

```python
df1 = data['Competition Name'].value_counts().reset_index()

plt.figure(figsize=(15,6))
sns.barplot(x="index", y="Competition Name", data=df1)
plt.title(" Number Of Teams Belonging To Each League")
plt.xlabel("Competition Name")
plt.ylabel("Number of Competition")
plt.xticks(rotation=80)
plt.show()
```

```python
df2 = data.groupby(["Club Name","Competition Name"]).mean().reset_index().drop("Unnamed: 0",axis=1)
df2.head()
```

```python
plt.figure(figsize=(22,7))
sns.barplot(y="Market Value Of Club In Millions(£)", x="Club Name", data=df2.sort_index())
plt.title("Relation between Club Name and Mean Market Value Of Club In Millions(£)")
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize=(22,7))
sns.barplot(y="Average Age Of Players", x="Club Name", data=df2.sort_index())
plt.title("Relation between Club Name and Average Age of Players")
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize=(22,7))
sns.barplot(y="Average Market Value Of Players In Millions(£)", hue="Squad Size",x="Competition Name", data=df2.sort_index())
plt.title("Relation between Average Market Value Of Players In Millions(£) and Squad Size")
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize=(22,7))
sns.barplot(y="Squad Size" ,x="Club Name", data=df2.sort_index())
plt.title("Relation between Average Club Name and Squad Size")
plt.xticks(rotation=80)
plt.show()
```
