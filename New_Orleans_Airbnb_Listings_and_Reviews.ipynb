```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', 500)
```

```python
listing_data = pd.read_csv("Datasets/New_Orleans_Airbnb_Listings_and_Reviews/new_orleans_airbnb_listings.csv")
review_data = pd.read_csv("Datasets/New_Orleans_Airbnb_Listings_and_Reviews/new_orleans_airbnb_reviews.csv")
```

```python
listing_data.head(3)
```

```python
listing_data.info()
```

# Threre are many **Null Data** in this DataSets so i will will droping all the Null Values

```python
listing_data.isnull().sum()
```

```python
listing_data = listing_data.dropna()
listing_data.head(2)
```

# Converting **Price** Column from str to float DataType

```python
listing_data["price"] = listing_data["price"].str.replace("$", "")
listing_data["price"] = listing_data["price"].str.replace(",", "")
listing_data["price"] = listing_data["price"].astype("float")
```

```python
listing_data.head(2)
```

# Slicing the **host_loction** Column and seperating to there respective Column ["host_state", "host_city", "host_country"]

```python
listing_data["count"] = listing_data["host_location"].str.split(",").apply(lambda x:len(x))
listing_data = listing_data[listing_data["count"] == 3]
listing_data["host_state"] = listing_data["host_location"].str.split(",").apply(lambda x:x[0])
listing_data["host_city"] = listing_data["host_location"].str.split(",").apply(lambda x:x[1])
listing_data["host_country"] = listing_data["host_location"].str.split(",").apply(lambda x:x[2])
listing_data.head(2)
```

# Analysing the Change in Price on the basis of host_location, review_rating_score and number of reviews 

```python
plt.figure(figsize=(20,7))
sns.barplot(x="host_location", y="price", data=listing_data)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.barplot(x="review_scores_rating", y="price", data=listing_data)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.barplot(y="number_of_reviews", x="review_scores_rating", data=listing_data)
plt.xticks(rotation=90)
plt.show()
```

```python
bins = [1, 1.5, 3, 4.5, 5]
ages = ["Poor", "Avereage", "Good", "Very Good"]
listing_data["review_scores_rating"] = pd.cut(listing_data["review_scores_rating"], bins=bins, labels=ages)
```

```python
plt.figure(figsize=(20,7))
sns.barplot(x="host_location", y="price", hue="review_scores_rating",  data=listing_data)
plt.xticks(rotation=90)
plt.show()
```

# Result of Analysing the Change in Price on the basis of host_location, review_rating_score and number of reviews.

#### 1) [West Hollywod, California, United States] and [Monrovia, California, United States] are the most expensive property but there  review_rating_score are also arround [4.5, 5] which is very good

#### 2) There are many property which got review_rating_score around [4.5, 5] and also comes in average price range of $250.
#### 3) Accoring to the number of rating the most comman ratting of a propert is around 4.5 which is pretty good
